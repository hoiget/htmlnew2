```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
iris=pd.read_csv('iris.csv')
```

```python
iris.drop(columns='Id',inplace=True)
```

```python
iris.head(10)
```

```python
def myCubeFuns(x:pd.DataFrame)->pd.DataFrame:
    return x*x*x
```

```python
x1=5
y1=7

x2=-5
y2=-7

d=(x1-x2)**2+(y1-y2)**2
print(np.sqrt(d))

dsquare=(x1**2-x2**2)**2+(y1**2-y2**2)**2
print(np.sqrt(dsquare))

dcube=(x1**3-x2**3)**2+(y1**3-y2**3)**2
print(np.sqrt(dcube))

```

```python
# Giả sử myCubeFuns đã định nghĩa và trả về DataFrame số
test_trans = iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
test_trans = myCubeFuns(test_trans)

# Lấy các cột số trong iris để tính toán
numeric_cols = iris.select_dtypes(include=['float64', 'int64']).columns

print("DATA FRAME DATA \n")
print("Dataframe \n", iris.head(5))
print("Dataframe after transforming \n", test_trans.head(5))
print("\n")

print("DESCRIBE FRAME DATA \n")
print("Dataframe \n", iris[numeric_cols].describe())
print("Dataframe after transforming \n", test_trans.describe())
print("\n")

print("SKEW REPORT \n")
print("iris skew \n", iris[numeric_cols].skew())
print("testdf skew \n", test_trans.skew())
print("\n")

print("KURTOSIS REPORT \n")
print("iris kurtosis \n", iris[numeric_cols].kurtosis())
print("testdf kurtosis \n", test_trans.kurtosis())
print("\n")
```

```python
X=processed_iris.iloc[:,0:4].values
y=processed_iris.iloc[:,4].values
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y= le.fit_transform(y)
```

```python
from sklearn.metrics import accuracy_score,precision_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
from sklearn.model_selection import KFold,train_test_split
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
gaussian =GaussianNB()
gaussian.fit(X_train,y_train)

Y_pred = gaussian.predict(X_test)
accuracy_nb=round(accuracy_score(y_test,Y_pred)* 100,2)
acc_gaussion = round(gaussian.score(X_train,y_train)* 100,2)

cm= confusion_matrix(y_test,Y_pred)
accuracy=accuracy_score(y_test,Y_pred)
precision=precision_score(y_test,Y_pred,average='micro')
recall=recall_score(y_test,Y_pred,average='micro')
f1=f1_score(y_test,Y_pred,average='micro')

print('Confusion matrix for Naive Bayes\n',cm)
print('accuracy_Naive Bayes: %.3f' %accuracy)
print('precision_Naive Bayes: %.3f' %precision)
print('recall_Naive Bayes: %.3f' %recall)
print('f1-score_Naive Bayes: %.3f' %f1)
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)

Y_pred = knn.predict(X_test)
accuracy_knn=round(accuracy_score(y_test,Y_pred)* 100,2)
acc_knn = round(knn.score(X_train,y_train)* 100,2)

cm= confusion_matrix(y_test,Y_pred)
accuracy=accuracy_score(y_test,Y_pred)
precision=precision_score(y_test,Y_pred,average='micro')
recall=recall_score(y_test,Y_pred,average='micro')
f1=f1_score(y_test,Y_pred,average='micro')

print('Confusion matrix for KNN\n',cm)
print('accuracy_KNN: %.3f' %accuracy)
print('precision_KNN: %.3f' %precision)
print('recall_KNN: %.3f' %recall)
print('f1-score_KNN: %.3f' %f1)
```

```python
plt.subplots(figsize=(20,5))
a_index = list(range(1,50))
accuracies = []  # dùng list lưu độ chính xác
x = range(1,50)

for i in a_index:
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    acc = accuracy_score(y_test, predictions)
    accuracies.append(acc)

plt.plot(a_index, accuracies, marker="*")
plt.xticks(x)
plt.show()
```

```python
decision_tree=DecisionTreeClassifier()
decision_tree.fit(X_train,y_train)

Y_pred = decision_tree.predict(X_test)
accuracy_dt=round(accuracy_score(y_test,Y_pred)* 100,2)
acc_decision_tree = round(decision_tree.score(X_train,y_train)* 100,2)

cm= confusion_matrix(y_test,Y_pred)
accuracy=accuracy_score(y_test,Y_pred)
precision=precision_score(y_test,Y_pred,average='micro')
recall=recall_score(y_test,Y_pred,average='micro')
f1=f1_score(y_test,Y_pred,average='micro')

print('Confusion matrix for DecisionTree\n',cm)
print('accuracy_DecisionTree: %.3f' %accuracy)
print('precision_DecisionTree: %.3f' %precision)
print('recall_DecisionTree: %.3f' %recall)
print('f1-score_DecisionTree: %.3f' %f1)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(15,10))
plot_tree(decision_tree.fit(X_train,y_train),filled=True)
plt.show()
```

```python
results = pd.DataFrame({
    'Model' : [
        'Naive Bayes',
        'KNN',
            'Decision Tree'
    ],
    'Score' : [
        acc_gaussion,
        acc_knn,
            acc_decision_tree
    ],
    "Accuracy_score" : [
        accuracy_nb,
        accuracy_knn,
            accuracy_dt
    ]
})

result_df=results.sort_values(by='Accuracy_score',ascending=False)
result_df=result_df.reset_index(drop=True)

print(result_df)
```
